```python
import numpy as np
import os
import datetime

white = True
erase = False
rows = 20
cols = rows
dim = 1080
radius = dim/(cols)
g = False
grid = np.zeros((rows, cols))
record = False
saving = False

timestamp = datetime.datetime.now().strftime('%y/%m/%d_%H-%M')
os.makedirs(f'grid_paint', exist_ok = True)

frame = 0

def setup():
    size (dim, dim)
    #no_loop()
    stroke_weight(3)

def draw():
    global grid
    global white
    global radius
    global g
    global timestamp
    global saving
    global record
    
    if white:
        background(0)
        fill(255)
        stroke(255)
    else:
        background(255)
        fill(0)
        stroke(0)
    w_c = width/grid.shape[1]
    h_c = height/grid.shape[0]
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            x = (j+0.5)*w_c
            y = (i+0.5)*h_c
            if grid[i][j] == 1:
                circle(x, y, radius)
                
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            x1 = (j+0.5)*w_c
            y1 = (i+0.5)*h_c
            if grid[i][j] == 1:
                neighbours = [(i, j+1), (i+1,j), (i+1,j-1), (i+1, j+1)]
                for i2, j2 in neighbours:
                    if i2 in range(0, grid.shape[0]) and j2 in range(0, grid.shape[1]):
                        if grid[i2][j2] == 1:
                            x2 = (j2+0.5)*w_c
                            y2 = (i2+0.5)*h_c
                            if abs((i2 + j2) - (i+j)) == 1:
                                rect_c(np.mean([x1,x2]), np.mean([y1, y2]), w_c)
                            else:
                                conj(np.mean([x1,x2]), np.mean([y1, y2]), radius/2, j2-j2)
    if g:
        for i in range(grid.shape[0]):
            stroke(255,0,0,50)
            line(0, i*w_c, width, i*w_c)
            line(i*w_c, 0, i*w_c, height)
    if saving:
        time_frame = datetime.datetime.now().strftime('%y%m%d%H%M')
        save_frame(f'grid_paint_{time_frame}.png')
        saving = False
    if record:
        save_frame(f'grid_paint/{time_stamp}/{frame_count}.png')

def mouse_dragged():
    global grid
    global erase
    global radius
    x_i = int(mouse_x // radius)
    y_i = int(mouse_y // radius)
    if erase:
        grid[y_i][x_i] = 0
    else:
        grid[y_i][x_i] = 1
    redraw()

def mouse_pressed():
    global grid
    global erase
    global radius
    x_i = int(mouse_x // radius)
    y_i = int(mouse_y // radius)
    if erase:
        grid[y_i][x_i] = 0
    else:
        grid[y_i][x_i] = 1
    redraw()
    
def key_pressed():
    global white
    global erase
    global g
    if key == 'w':
        white = not white
        redraw()
    if key == 'm':
        erase = not erase
    if key == 'g':
        g = not g
        redraw()
    if key == 'r':
        record = not record
        os.makedirs(f'grid_paint/{timestamp}', exist_ok = True)
    if key == 's':
        saving = True

def rect_c(x, y, size):
    rect_mode(CENTER)
    square(x, y, size)
    
def conj(xc, yc, radius, post):
    k = 4/3*(sqrt(2)-1)
    begin_shape()
    vertex(xc, yc-radius)
    bezier_vertex(
            xc , yc-radius + k * radius,
            xc + radius - k * radius, yc,
            xc+radius, yc)
    bezier_vertex(
            xc+radius - k * radius, yc,
            xc, yc + radius - k * radius,
            xc, yc + radius)
    bezier_vertex(
            xc, yc+ radius - k * radius,
            xc-radius + k * radius, yc,
            xc-radius, yc)
    bezier_vertex(
            xc-radius + k * radius, yc,
            xc , yc-radius + k * radius,
            xc, yc-radius)
    end_shape(CLOSE)
    

run_sketch()
```
